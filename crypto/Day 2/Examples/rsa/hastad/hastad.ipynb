```python
from Crypto.Util.number import getPrime, inverse

moduli = []
for _ in range(3):
    p = getPrime(128)
    q = getPrime(128)
    N = p*q
    moduli.append(N)

e = 3
#d = inverse(e,(p-1)*(q-1))
```

```python
from Crypto.Util.number import bytes_to_long as btl
from Crypto.Util.number import bytes_to_long as ltb

m_1 = btl(b'Here\'s my secret message')

for n in moduli:
    assert pow(m_1,e) > n
    
ciphertexts = []
for n in moduli:
    ciphertexts.append(pow(m_1,3,n))
```

```python
from sage.all import *
for c in ciphertexts:
    try:
        val = c.nth_root(3)
        print(ltb(val))
    except:
        pass
```

### Attack
The prior code block is to clear any doubts that with a public key value of three that we can just obtain the original message by taking the cube root. Remmeber, said approach only works in the case that we do not wrap around the modulus. However, notice I used an assertion to check that for each message value that we do exceed the modulus. 

So how do obtain the messages?

Hastad's Broadcast Attack

```python
x = CRT(ciphertexts,moduli)
root = x.nth_root(3)
```
